**Efetividade de Duas Estratégias de Ensino**

Imagine que uma escola esteja avaliando a eficácia de duas estratégias de ensino de matemática para alunos do ensino médio. Eles querem determinar se há uma diferença significativa no desempenho médio dos alunos entre as duas estratégias.

# **Hipóteses:**

* Hipótese nula (H0): A média das notas dos alunos na estratégia A é igual à média das notas dos alunos na estratégia B.
* Hipótese alternativa (H1): A média das notas na Estratégia B é maior do que a média das notas na Estratégia A.

# **Dados:**

* Amostra da Estratégia A: Notas de 50 alunos que receberam a Estratégia A.
* Amostra da Estratégia B: Notas de 50 alunos que receberam a Estratégia B.

Usaremos um teste Z para comparar as médias das notas entre as duas amostras.

Se o p-valor do teste Z for menor que um nível de significância pré-determinado (por exemplo, α = 0.05), rejeitamos a hipótese nula e concluímos que há uma diferença significativa nas médias das notas entre as duas estratégias de ensino.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Os dados são criados a seguir:

```python
# Definindo médias e desvios padrão para as notas nas duas estratégias
media_estrategia_A = 70
desvio_padrao_estrategia_A = 10

media_estrategia_B = 75
desvio_padrao_estrategia_B = 12

# Gerando as amostras de notas para cada estratégia de ensino da nossa base
np.random.seed(0)  # Para garantir a reprodutibilidade dos resultados
amostra_estrategia_A = np.random.normal(loc=media_estrategia_A, scale=desvio_padrao_estrategia_A, size=50)
amostra_estrategia_B = np.random.normal(loc=media_estrategia_B, scale=desvio_padrao_estrategia_B, size=50)

print("Notas da Estratégia A:", amostra_estrategia_A[:5])
print("Notas da Estratégia B:", amostra_estrategia_B[:5])
```

# 1) De acordo com as informações analisadas o nosso teste é unilateral á direita, esquerda ou bicaudal? Justifique.




```python
# Sua justificativa aqui
# Inicialmente é possível verificar que as notas da estratégia A possuem uma maior nota, sendo assim é possível que seja um teste unilateral à direita
# Justamente pela forma que consta a fórmula que indica isto (μ_A > μ_B) após tirar a média e a variância ficará mais claro.
```

# 2) Calcule as médias para as duas amostragens e as variâncias. Quais insights você pode retirar comparando os dados?

```python
#Seu código aqui
# Para verificar a variância utilizarei a função 'var'
media_A = np.mean(amostra_estrategia_A)
var_A = np.var(amostra_estrategia_A, ddof=1)  # variância amostral

media_B = np.mean(amostra_estrategia_B)
var_B = np.var(amostra_estrategia_B, ddof=1)

print("Média A:", media_A)
print("Variância A:", var_A)
print("Média B:", media_B)
print("Variância B:", var_B)
```

###Aqui podemos ver que inicialmente tiramos a conclusão que a estratégia A era qual indicava ser maior, porém por meio do print da média e da variância foi possível perceber que na verdade a estratégia B é maior com uma variância mais concentrada. Continuará sendo um teste unilateral a esquerda porém de uma forma 'invertida' pois será desta forma: (μ_B > μ_A)

# 3) Imprima os resultados da estatística do teste Z, p value e indique se rejeitaremos ou não a hipótese nula. Justifique sua resposta.

```python
#Seu código aqui
from scipy.stats import norm

# Regerando as amostras com seed para garantir consistência
np.random.seed(0)
amostra_estrategia_A = np.random.normal(loc=70, scale=10, size=50)
amostra_estrategia_B = np.random.normal(loc=75, scale=12, size=50)

# Médias e tamanhos
media_A = np.mean(amostra_estrategia_A)
media_B = np.mean(amostra_estrategia_B)
n_A = len(amostra_estrategia_A)
n_B = len(amostra_estrategia_B)

# Variâncias amostrais
var_A = np.var(amostra_estrategia_A, ddof=1)
var_B = np.var(amostra_estrategia_B, ddof=1)

# Estatística do teste Z (para médias de duas populações independentes)
z_stat = (media_B - media_A) / np.sqrt(var_A/n_A + var_B/n_B)

# Para um teste unilateral à direita, o p-value é 1 - P(Z < z_stat)
p_value = 1 - norm.cdf(z_stat)

# Nível de significância
alpha = 0.05

print(f"Estatística Z: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Nível de significância : {alpha}")
print("-" * 30)

# Verificação da hipótese nula
if p_value < alpha:
    print("O p-value é menor. Rejeitamos a hipótese nula.")
    print("Isso sugere que a média da estratégia B é estatisticamente maior que a da A.")
else:
    print("O p-value é maior ou igual. Não rejeitamos a hipótese nula.")
    print("Não há evidência estatística suficiente para dizer que a média da estratégia B é maior que a da A.")
```

###O próprio método utilizado já nos indica que não reijeitaremos a hipótese nula, pois nosso p-value nos indica que o resultado apontado seja por acaso, mesmo que a hipótese nula seja verdadeira que neste caso se trata de 6%. Como nossa probabilidade é maior do que nosso limite de 5%. não posso afirmar que temos evidências suficientes para rejeitar a hipótese nula, desta forma não se trata de um evento raro, só poderia ser uma hipótese nula caso estivesse abaixo do nosso nível de significância.

# 4) Para finalizar monte o gráfico da distribuição da estatística do nosso teste Z e explique o que pode ser observado através dele.

```python
#Seu código aqui
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Valores fornecidos
z_stat = 1.5267
p_value = 0.0634
alpha = 0.05

# Nível de significância
alpha = 0.05

# Gera os valores do eixo x para a distribuição normal padrão
x = np.linspace(-3, 3, 1000)

# Calcula a função densidade de probabilidade (PDF) para esses valores
pdf = norm.pdf(x, 0, 1)

# Cria o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', lw=2, label='Distribuição Normal Padrão')

# Preenche a área do p-value (área da cauda direita)
x_p_value = np.linspace(z_stat, 3, 1000)
plt.fill_between(x_p_value, norm.pdf(x_p_value), color='red', alpha=0.5, label=f'P-value (Área vermelha) = {p_value:.4f}')

# Preenche a área de rejeição (área da cauda direita a partir do Z crítico)
z_critico = norm.ppf(1 - alpha)
x_critico = np.linspace(z_critico, 3, 1000)
plt.fill_between(x_critico, norm.pdf(x_critico), color='gray', alpha=0.3, label=f'Área de Rejeição (Z crítico = {z_critico:.2f})')

# Adiciona a linha da estatística Z observada
plt.axvline(x=z_stat, color='r', linestyle='--', lw=2, label=f'Estatística Z = {z_stat:.4f}')

# Adiciona a linha do Z crítico
plt.axvline(x=z_critico, color='k', linestyle=':', lw=2, label=f'Z Crítico (alpha={alpha}) = {z_critico:.2f}')

# Adiciona legendas e títulos
plt.title('Teste Unilateral à Direita (Comparação de Z-stat e Z-crítico)')
plt.xlabel('Estatística Z')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

##O que o gráfico mostra?
###O gráfico acima representa a distribuição normal padrão (a curva em azul), que é a base para o nosso teste Z. Ele nos ajuda a visualizar a nossa decisão.

####A linha pontilhada preta (Z Crítico): Esta linha marca o limite da nossa área de rejeição. Com um
####alpha=0.05, qualquer valor Z que caia à direita dessa linha nos levaria a rejeitar a hipótese nula. O valor exato para esse limite é 1.64, que é o Z-crítico para um teste unilateral à direita. A área cinza representa a área de rejeição.

####A linha tracejada vermelha (Estatística Z): Esta linha representa o valor da estatística Z que calculamos com nossos dados: 1.5267. Ela mostra onde nosso resultado se encontra na distribuição.

####A área vermelha (p-value): Esta área representa o p-value, que é a probabilidade de obtermos um resultado tão extremo (ou mais extremo) quanto o nosso Z-stat, se a hipótese nula fosse verdadeira. A área vermelha é o p-value (0.0634).

###Conclusão Visual
####Olhando para o gráfico, você pode ver que a linha vermelha (nosso Z-stat) está à esquerda da linha pontilhada preta (o Z-crítico). Isso significa que o nosso resultado não entrou na área de rejeição.

####A área vermelha (p-value) é um pouco maior que a área cinza (alpha). Como o p-value é maior que o nosso nível de significância, a decisão é não rejeitar a hipótese nula. Não há evidência estatística forte o suficiente para afirmar que a Estratégia B é superior à Estratégia A.
